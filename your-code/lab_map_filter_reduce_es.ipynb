#### El siguiente paso es crear una función que eliminará las referencias.

Haremos esto dividiendo la cadena en el carácter `{` y manteniendo solo la parte antes de este carácter. Escribe tu función a continuación.

```python
def referencia(x:list) -> list:
    '''
    Entrada (list of strings): Una lista de cadena de carácteres en la que se quiere eliminar las referencias.
    Salida (list of strings): La cadena con las referencias eliminadas

    Ejemplo:
    Entrada: 'the{7}'
    Salida: 'the'
    '''
    new_list = []
    for palabras in x:
      # Se separan todas las palabras si tienen {
      parts = palabras.split("{")
      # Nos quedamos con la parte de la palabra antes de {, o sea la que no contiene la referencia
      new_list.append(parts[0])
    return (new_list)
```

```python
# Se aplica la función a la parte de los datos que son parte del libro en sí (a partir de la palabra 568)
# Se muestra el resultado que se observa sin las referencias tipo {número}
referencia(prophet[568:])
```

Ahora que tenemos nuestra función, usa la función `map()` para aplicar esta función a nuestro libro, El Profeta. Devuelve la lista resultante a una nueva lista llamada `prophet_reference`.

```python
# Construimos la función para mapear el string, no la lista entera.
def referencia_map(x: str) -> str:
    '''
    Entrada: Una cadena de carácteres en la que se quiere eliminar las referencias
    Salida: La cadena con las referencias eliminadas

    Ejemplo:
    Entrada: 'the{7}'
    Salida: 'the'
    '''
    # Se separan todas las palabras si tienen {
    parts = x.split("{")
    # Nos quedamos con la parte de la palabra antes de {, o sea la que no contiene la referencia
    return parts[0]
```

```python
prophet_reference = list(map(referencia_map, prophet[568:]))
""" map(fun, iter)
fun: Es una función a la cual map le pasa cada elemento del iterable dado.
iter: Es iterable lo que se va a mapear."""
prophet_reference
```

Otra cosa que podrías haber notado es que algunas palabras contienen un salto de línea. Escribamos una función para dividir esas palabras. Nuestra función devolverá la cadena dividida en el carácter `\n`. Escribe tu función en la celda de abajo.

```python
def salto_de_linea(x:list)->list:
  '''
  Entrada (list of str): Una lista de cadenas en la que hay cadenas que contienen 1 o varios saltos de línea (\n) que se desean dividir.
  Salida (list of str): Una lista de cadenas divididas en el carácter de salto de línea (\n)

  Ejemplo:
  Entrada: 'the\nbeloved'
  Salida: ['the', 'beloved']

  Entrada: 'the\n\nbeloved'
  Salida: ['the', 'beloved']
  '''
  new_list = []
  for palabras in x:
    # Cuando haya saltos de línea se separan las palabras por \n
    if "\n" in palabras:
      parts = palabras.split("\n")
      # Se añaden a la lista las dos partes asegurando que en la segunda parte quedan eliminados todos los saltos de línea addicionales
      new_list.append(parts[0])
      new_list.append(parts[1].replace('\n', ''))
    # Cuando no haya saltos de línea se añade la palabra que había
    else:
      new_list.append(palabras)
  return (new_list)
```

Aplica la función `salto_de_linea` a la lista `prophet_reference`. Nombra la nueva lista `prophet_line`.

```python
prophet_line = salto_de_linea(prophet_reference)
prophet_line
```

```python
# Otra forma de hacerlo es construimos la función para mapear el string, no la lista entera y devolver una lista en cada palabra que hay salto de línea.
def salto_de_linea(x:str)->list:
  '''
  Entrada (str): Una cadenas que contienen 1 o varios saltos de línea (\n) que se desean dividir.
  Salida (list of str): Una lista con las cadenas divididas por el salto de línea (\n)

  Ejemplo:
  Entrada: 'the\nbeloved'
  Salida: ['the', 'beloved']

  Entrada: 'the\n\nbeloved'
  Salida: ['the', 'beloved']
  '''
  # Si hay salto de línea, se separan y retornan las listas de palabras separadas.
  if "\n" in x:
    # Se separa la palabra según el salto de línea
    parts = x.split("\n")
    # Se retorna una lista con cada parte separada asegurando que en la segunda parte quedan eliminados todos los saltos de línea addicionales.
    return([parts[0], parts[1].replace('\n', '')])
  # Si no hay salto de línea se devuelve la misma palabra.
  else:
    return([x])
```

```python
prophet_line = list(map(salto_de_linea,prophet_reference))
prophet_line
```

```python
prophet_flat = [i for sub in prophet_line for i in sub]
prophet_flat
```

# Desafío 2 - Filtering

Al imprimir algunas palabras del libro, vemos que hay palabras que tal vez no queramos conservar si elegimos analizar el corpus de texto. A continuación, se muestra una lista de palabras que nos gustaría eliminar. Crea una función que devuelva falso si contiene una palabra de la lista de palabras especificadas y verdadero de lo contrario.

```python
def filtro_palabra(x:str) -> bool:
    '''
    Entrada: Una cadena
    Salida: Verdadero si la palabra no es alguna de estas ['and', 'the', 'a', 'an']
              y Falso si la palabra si es alguna de estas ['and', 'the', 'a', 'an']

    Ejemplo:
    Entrada: 'and'
    Salida: FALSE

    Entrada: 'John'
    Salida: TRUE
    '''
    lista_palabras = ['and', 'the', 'a', 'an']
    if x in lista_palabras:
      return False
    else:
      return True
```

Usa la función `filter()` para filtrar las palabras especificadas en la función `filtro_palabra()`. Guarda la lista filtrada en la variable `prophet_filter`.


```python
# Se filtran las palabras definidas en la función filtro_palabra
prophet_filter = list(filter(filtro_palabra, prophet_flat))
'''filter(fun, iter)
Entradas:
  fun: Es la función que establece "True" o "False" en cada elemento del iterable.
  iter: Es iterable que se desea filtrar.
Salida:
  iterable filtrado con los "True" de la función'''
prophet_filter
```

# Desafío 3 - Reducing

#### Ahora que hemos limpiado significativamente nuestro corpus de texto, usemos la función `reduce()` para volver a unir las palabras en una cadena larga separada por espacios.

Comenzaremos escribiendo una función que toma dos cadenas y las concatena juntas con un espacio entre las dos cadenas.


```python
def concatenar_espacio(a:str, b:str) -> str:
  '''
  Entrada (str): Dos cadenas.
  Salida (str): Una única cadena separada por un espacio

  Ejemplo:
  Entrada: 'John', 'Smith'
  Salida: 'John Smith'
  '''
  return a + ' ' + b
```

Usa la función de arriba para reducir el corpus de texto en la lista `prophet_filter` a una única cadena. Asigna esta nueva cadena a la variable `prophet_string`.

```python
prophet_string = reduce(concatenar_espacio, prophet_filter)
'''reduce(función, iterable)
Se aplica una function a todos los elementos del iterable de la siguiente manera:
1) Se cogen los dos primeros elementos, se aplica la función con esos y se obtiene un resultado.
2) Se aplica la misma función con el resultado obtenido y el siguiente elemento (el tercero) y se obtiene otro resultado.
3) El proceso continua con todos los elementos del iterable, cogiendo siempre el resultado obtenido y el siguiente elemento.
4) El resultado final es el que se retorna.'''
prophet_string
```
